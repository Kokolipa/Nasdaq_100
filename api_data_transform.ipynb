## Create Metadata .csv file

```python
# Import csv file of metadata created in API call into a dataframe and preview column names
overview = pd.read_csv("stockdata/overview_nasdaq.csv", encoding='UTF-8')
overview.columns.values.tolist()
```

```python
# Preview dataframe
overview.head()
```

```python
# Create new column with unique ID for each ticker
overview["ID"] = range(len(overview))
```

```python
# Preview
overview.head()
```

```python
# Mobe the ID column to the beginning of dataframe
id_column = overview.pop("ID")
overview.insert(0, "ID", id_column)
```

```python
overview.head()
```

```python
# Preview column names
overview.columns.values.tolist()
```

```python
# Create new dataframe with selected columns only
metadata_start = overview[["ID", "Symbol", "AssetType", "Name", "Description", "Currency", "Country", "Sector", "Industry", "MarketCapitalization", 
                  "EBITDA", "BookValue", "DividendPerShare", "DividendYield"]]
```

```python
# Preview new dataframe
metadata_start.head()
```

```python
# Rename columns
metadata = metadata_start.rename(columns={"ID": "ticker_id", "Symbol": "symbol", "AssetType": "asset_type", "Name": "name", "Description": "description", "Currency": "currency",
                         "Country": "country", "Sector": "sector", "Industry": "industry",
                         "MarketCapitalization": "market_capitalization", "EBITDA": "ebitda",
                         "BookValue": "book_value", "DividendPerShare": "dividend_per_share",
                         "DividendYield": "dividend_yield"})
```

```python
# Check data types
metadata.dtypes
```

```python
# Export cleaned metadata dataframe to csv file
metadata.to_csv("stockdata/cleaned_metadata.csv", encoding='UTF-8', index=False)
```
